```python
df.columns
```

```python
df_knn = df[['ram','battery_power','int_memory','mobile_wt',
             'px_width','px_height','price_range']]
df_knn.shape
```

```python
df_knn.sample(7)
```

```python
x = df_knn.drop('price_range',axis=1)
y = df_knn['price_range']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,
                                                random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
print(type(x_train_sc))
print(type(x_test_sc))
print()
```

```python
knn = KNeighborsClassifier(n_neighbors=31)
knn.fit(x_train_sc,y_train)
```

```python
print('Train score',knn.score(x_train_sc,y_train))
print('Test score',knn.score(x_test_sc,y_test))
```

```python
ypred_knn = knn.predict(x_test_sc)
print(ypred_knn)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def gen_matrix(y_test,y_pred):
    cm = confusion_matrix(y_test,y_pred)
    print(cm)
    print(classification_report(y_test,y_pred))
```

```python
gen_matrix(y_test,ypred_knn)
```

#### Accuracy vs No of Neighbors

```python
neighbors = list(range(3,51,2))
knn_acc = []
for i in neighbors:  # i=3,5...49
    m = KNeighborsClassifier(n_neighbors=i)
    m.fit(x_train_sc,y_train)
    ypred = m.predict(x_test_sc)
    knn_acc.append(accuracy_score(y_test,ypred))
#     knn_acc.append(m.score(y_test,ypred))
```

```python
plt.plot(neighbors,knn_acc,color='maroon',label='Acc',marker='o')
plt.title('Accuracy vs No of Neighbors - KNN')
plt.xlabel('No of Neighbors')
plt.ylabel('Test Accuracy')
plt.legend()
plt.grid()
plt.show()
```
